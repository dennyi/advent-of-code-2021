```python
file1 = open('input.txt', 'r')
lines = file1.read().split('\n')

map = np.zeros((len(lines), len(lines[0])), dtype=int)

for i in range(0, len(lines)):
    map[i, :] = list(lines[i])

def is_lowest_point(row, col):
    point = map[row, col]
    if((row+1 < map.shape[0] and map[row+1, col] <= point) or
    (row-1 >= 0 and map[row-1, col] <= point) or
    (col+1 < map.shape[1] and map[row, col+1] <= point) or
    (col-1 >= 0 and map[row, col-1] <= point)):
        return False
    return True


lowest_points = []
for row in range(0, map.shape[0]):
    for col in range(0, map.shape[1]):
        if(is_lowest_point(row, col)):
            lowest_points.append((row,col))



def get_adjanced_points(row, col):
    adjanced_points = []
    if(row+1 < map.shape[0]):
        adjanced_points.append((row+1, col))
    if(row-1 >= 0 ):
        adjanced_points.append((row-1, col))
    if(col+1 < map.shape[1]):
        adjanced_points.append((row, col+1))
    if(col-1 >= 0):
        adjanced_points.append((row, col-1))
    return adjanced_points

def get_lowest_adjanced_points(row, col):
    lowest_value = map[row, col]
    lowest_points = [(row, col)]
    adjanced_points = get_adjanced_points(row, col)
    for point in adjanced_points:
        if(map[point[0], point[1]] < lowest_value and map[point[0], point[1]] != 9):
            lowest_points = [point]
            lowest_value = map[point[0], point[1]]
        elif(map[point[0], point[1]] == lowest_value and map[point[0], point[1]] != 9):
            lowest_points.append(point)

    return lowest_points



def get_points_flowing_to_here(row, col):
    center_point = (row, col)
    points = []
    
    adjanced_points = get_adjanced_points(row, col)
    for point in adjanced_points:
        if(center_point in get_lowest_adjanced_points(point[0], point[1]) and map[point[0],point[1]] != 9):
            points.append(point)
    return points
    

def get_low_point_basin(row, col):
    final_basin = [(row, col)]
    points = []
    points = get_points_flowing_to_here(row, col)
    while(len(points) > 0):
        new_points = []
        for i in range(0, len(points)):
            point = points[i]
            final_basin.append(point)
            calculated_points = get_points_flowing_to_here(point[0], point[1])
            if(len(calculated_points) > 0):
                new_points = calculated_points + new_points
        points = new_points
    filtered_basin = []
    [filtered_basin.append(x) for x in final_basin if x not in filtered_basin]
    return filtered_basin
    

basins = []

print(get_low_point_basin(4,6), len(get_low_point_basin(4,6)))

for low_point in lowest_points:
    basins.append(get_low_point_basin(low_point[0], low_point[1]))

basins.sort(key=len, reverse=True)

largest_basins_size = 1
for i in range(0, 3):
    largest_basins_size *= len(basins[i])

print(largest_basins_size)
```
